```python
spark.conf.set(
  "fs.azure.account.key.salesviewacc.dfs.core.windows.net",
  "<accesskey>"
)
```

```python
product = spark.read.format("delta").load(
  "abfss://silver@salesviewacc.dfs.core.windows.net/sales_view/product"
)

store = spark.read.format("delta").load(
  "abfss://silver@salesviewacc.dfs.core.windows.net/sales_view/store"
)

sales = spark.read.format("delta").load(
  "abfss://silver@salesviewacc.dfs.core.windows.net/sales_view/customer_sales"
)
```

```python
from pyspark.sql.functions import col
df1 = product.join(store, col("store_id") == col("store__i_d"), "inner")

```

```python
from pyspark.sql.functions import col
final_df = sales.join(df1, col("product__id") == col("product__id"), "inner")

```

```python
final_df = final_df.select(
    "order_date",
    "category_id",
    "city",
    "customer_i_d",
    "order_i_d",
    "product__id",
    "profit",
    "region",
    "sales",
    "segment",
    "ship_date",
    "ship_mode",
    "latitude",
    "longitude",
    "store_name",
    "location",
    "manager_name",
    "product_name",
    "price",
    "stock_quantity",
    "image_url"
)
```

```python
final_df.write.format("delta") \
  .mode("overwrite") \
  .save("abfss://gold@salesviewacc.dfs.core.windows.net/sales_view/storeproductsalesanalysis")
```
